FINDING THE BEST AND WORST MOVIES

```python
def best_worst(n,by,ascending=False,min_bud=0,min_votes=0):
    df2 = df_best.loc[(df_best.Budget>=min_bud) & (df_best.Votes>=min_votes),["",by]].sort_values(by=by,ascending = ascending).head(n).copy()
    return HTML(df2.to_html(escape=False))
```

```python
# Top 5 Highest Revenue
best_worst(5,"Revenue")
```

```python
df2 = df_best.loc[(df_best.Budget>=0) & (df_best.Votes>=0),["","Revenue"]].sort_values(by="Revenue",ascending = False).head(5).copy()
HTML(df2.to_html(escape=False))
```

```python
#Top 5 budgets
best_worst(n=5,by="Budget")
```

```python
#Top 5 Highest profit
best_worst(5,"Profit")
```

```python
#top 5 lowest profit
best_worst(5,"Profit",ascending=True)
```

```python
# top 5 Highest ROI
best_worst(5,"ROI",min_bud=5)
# top 5 lowest ROI
best_worst(5,"ROI",ascending=True,min_bud=100)
# top 5 votes
best_worst(5,"Votes")
# top 5 average Rating
best_worst(5,"Average Rating",min_votes=10)
# top 5 popular movies
best_worst(5,"Popularity")
```

```python
df.set_index("id",inplace=True)
```

```python
df.cast
```

```python
act = df.cast.str.split("|",expand=True)
act
```

```python
act = act.stack().reset_index(level=1,drop=True).to_frame()
act
```

```python
act.columns = ["Actors"]
```

```python
act
```

```python
act = act.merge(df[['title',"revenue_musd","vote_average","popularity"]],how="left",left_index=True,right_index=True)
```

```python
act
```

```python
act.Actors.value_counts().head(20)
```

```python
agg = act.groupby("Actors").agg(Total_Revenue = ("revenue_musd","sum"),
                               Mean_Revenue = ("revenue_musd","mean"),
                               Mean_pop = ("popularity","mean"),
                               Total_movies = ("Actors","count"))
agg
```

```python
agg.nlargest(20,"Total_movies")
```

```python
agg[agg.Total_movies>=10].nlargest(10,"Mean_Revenue")
```
